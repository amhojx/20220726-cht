### 練習：

請用 folium 將上面 100 個站點在地圖上標出來！

````python
place = "台北市"
````

````python
gps = geocoder.osm(place).latlng
````

````python
m = folium.Map(gps, zoom_start=16)
````

````python
folium.Marker(gps, popup=place, icon=folium.Icon(color="red")).add_to(m)
````

````python
m
````

````python
for i in range(100):
  x = data[i]
  # print(x['sna'], "==>", x['lat'], ",", x['lng'])
  folium.Marker((x['lat'], x['lng']), popup=x['sna']).add_to(m)
````

````python
m
````

````python
for x in data:
  folium.Marker((x['lat'], x['lng']), popup=x['sna']).add_to(m)
````
